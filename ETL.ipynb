# 1 ESTRACCION AQUI TEMBIEN VOY A DESANIDAR

Al revisar el diccionario de datoa  de Steam pude ver que dos de los 'user_reviews'  e 'users_items' podrian no ser jsom estandar, lo notamos por las comillas simples estoces usamos la funcion ast.literal_eval(line.decode('utf-8') para que se permita su extracion  y luego podamos estadarizarlos 

```python
import gzip
import json
import pandas as pd
import ast
# Función para cargar datos desde archivos gzip estee a
data = []
with gzip.open('steam_games.json.gz', 'rb') as f:
    for line in f:
        data.append(json.loads(line))
steam_games = pd.DataFrame(data)
f.close()


data = []
with gzip.open('user_reviews.json.gz', 'rb') as f:
    for line in f:
        data.append(ast.literal_eval(line.decode('utf-8')))
user_reviews = pd.DataFrame(data)
f.close()


data = []
with gzip.open('users_items.json.gz', 'rb') as f:
    for line in f:
        data.append(ast.literal_eval(line.decode('utf-8')))
users_items = pd.DataFrame(data)
f.close()



# Mostrar información básica de los DataFrames
print("Información básica del DataFrame 'steam_games':")
print(steam_games.info())
print("\nPrimeras filas del DataFrame 'steam_games':")
print(steam_games.head())

print("\nInformación básica del DataFrame 'user_reviews':")
print(user_reviews.info())
print("\nPrimeras filas del DataFrame 'user_reviews':")
print(user_reviews.head())

print("\nInformación básica del DataFrame 'users_items':")
print(users_items.info())
print("\nPrimeras filas del DataFrame 'users_items':")
print(users_items.head())
```
